```python
import matplotlib.pyplot as plt
import numpy as np

Categories=['Category A','Category B','Category C','Category D','Category E']
Values=[23,56,41,62,19]

Data= [12, 17, 21, 18, 14, 13, 16, 9, 12, 15, 19, 11, 14, 16, 20, 18, 15, 13, 16, 11, 10]

Labels= ['Apple', 'Banana', 'Orange', 'Mango']
Sizes=[30, 25, 15, 30]

X_values=[1,2,3,4,5]
Y_values=[2,5,3,6,4]

plt.figure(figsize=(20,4))
plt.subplots_adjust(bottom=0.1,right=0.8,top=2.2)

plt.subplot(2,2,1)
plt.title('Bar chart',color='g')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.bar(Categories,Values)


plt.subplot(2,2,2)
plt.title('Histogram',color='r')
plt.xlabel('Data')
plt.ylabel('values')
plt.hist(Data)


plt.subplot(2,2,3)
plt.title('Pie chart',color='b')
plt.pie(Sizes,labels=Labels)


plt.subplot(2,2,4)
plt.title('Scatter Plot',color='m')
plt.xlabel('X_values')
plt.ylabel('Y_values')
plt.scatter(X_values,Y_values)


plt.show()
```
